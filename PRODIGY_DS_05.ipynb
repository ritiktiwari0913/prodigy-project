```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your DataFrame (assuming it's already loaded as data)


# 1. Distribution of accidents by day of the week
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='Day_of_week', order=data['Day_of_week'].value_counts().index)
plt.title('Distribution of Accidents by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.show()

# 2. Distribution of accidents by age band of the driver
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='Age_band_of_driver', order=data['Age_band_of_driver'].value_counts().index)
plt.title('Distribution of Accidents by Age Band of Driver')
plt.xlabel('Age Band of Driver')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.show()

# 3. Correlation heatmap
# Ensure numeric columns
numeric_columns = data.select_dtypes(include=['number']).columns
corr_matrix = data[numeric_columns].corr()

plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
data = pd.read_csv("task5.csv")

# Fill missing values with mode (most frequent value) for each column
for column in data.columns:
    mode_value = data[column].mode()[0]
    data[column].fillna(mode_value, inplace=True)


# Additional plots

# 4. Distribution of accidents by type of vehicle
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='Type_of_vehicle', order=data['Type_of_vehicle'].value_counts().index)
plt.title('Distribution of Accidents by Type of Vehicle')
plt.xlabel('Type of Vehicle')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.show()

# 5. Distribution of accidents by weather conditions
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='Weather_conditions', order=data['Weather_conditions'].value_counts().index)
plt.title('Distribution of Accidents by Weather Conditions')
plt.xlabel('Weather Conditions')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.show()

# 6. Distribution of accidents by light conditions
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='Light_conditions', order=data['Light_conditions'].value_counts().index)
plt.title('Distribution of Accidents by Light Conditions')
plt.xlabel('Light Conditions')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.show()


```

```python
plt.figure(figsize=(10, 6))
filtered_data = data[data['Sex_of_driver'] != 'Unknown']
sns.countplot(data=data, x='Sex_of_driver', order=filtered_data['Sex_of_driver'].value_counts().index)
plt.title('Distribution of Accidents by Sex of Driver')
plt.xlabel('Sex of Driver')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.show()
```

```python
sns.pairplot(data[['Driving_experience', 'Number_of_vehicles_involved', 'Number_of_casualties']])
plt.suptitle('Pair Plot of Driving Experience and Other Variables', y=1.05)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=data, x='Road_surface_type', y='Number_of_casualties')
plt.title('Number of Casualties by Road Surface Type')
plt.xlabel('Road Surface Type')
plt.ylabel('Number of Casualties')
plt.xticks(rotation=45)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Count occurrences of each cause
cause_counts = data['Cause_of_accident'].value_counts()

plt.figure(figsize=(10, 10))
plt.pie(cause_counts, labels=cause_counts.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired(range(len(cause_counts))))
plt.title('Proportion of Accidents by Cause')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
sns.countplot(data=data, x='Cause_of_accident', order=data['Cause_of_accident'].value_counts().index)
plt.title('Number of Accidents by Cause')
plt.xlabel('Cause of Accident')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.show()
```

```python
# Convert 'Time' to a categorical variable (e.g., time of day)
data['Time_of_day'] = pd.cut(pd.to_datetime(data['Time']).dt.hour, bins=[0, 6, 12, 18, 24], labels=['Night', 'Morning', 'Afternoon', 'Evening'])

plt.figure(figsize=(14, 7))
sns.countplot(data=data, x='Cause_of_accident', hue='Time_of_day', order=data['Cause_of_accident'].value_counts().index)
plt.title('Number of Accidents by Cause and Time of Day')
plt.xlabel('Cause of Accident')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.legend(title='Time of Day')
plt.show()
```
